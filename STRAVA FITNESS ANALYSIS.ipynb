# Sleep and Activity Analysis  

In this analysis, I examined the relationship between **physical activity (steps)** and **sleep patterns** using the merged dataset of activity and sleep data. The analysis is divided into three parts:  


## 1. Correlation between Total Minutes Asleep and Total Steps  

I calculated the Pearson correlation coefficient to measure the relationship between the **total number of steps taken in a day** and the **total minutes of sleep**.  

- A **positive correlation** would mean that higher activity levels (more steps) are associated with longer sleep duration.  
- A **negative correlation** would mean that higher activity levels are linked with shorter sleep duration.  
- A correlation value close to **0** would indicate little or no relationship between the two.  

This helps in understanding whether being more active influences sleep patterns.  


## 2. Scatter Plot: Total Steps vs. Total Minutes Asleep  

To visualize this relationship, I plotted the total steps (x-axis) against the total minutes asleep (y-axis).  

- Each point represents one day of data for a user.  
- If the points show an upward trend, it suggests that more activity leads to more sleep.  
- If the points scatter randomly without a clear pattern, it indicates that steps and sleep are not strongly related.  

This visualization supports the correlation analysis and provides an intuitive view of the relationship.  

## 3. Average Sleep Duration per Weekday  

I also analyzed **average sleep duration across different days of the week** to see if people tend to sleep differently on weekdays versus weekends.  

- The day of the week was extracted from the activity date.  
- Then, I calculated the **average sleep duration** for each weekday.  
- A bar chart was used to display these averages clearly.  

This helped identify patterns such as:  
- Do people sleep longer on weekends compared to weekdays?  
- Is sleep duration lower at the start or end of the work week (e.g., Monday or Friday)?  


## Key Takeaways  

- The **correlation analysis** quantifies the relationship between daily activity and sleep.  
- The **scatter plot** provides a visual check for this relationship.  
- The **weekday analysis** shows behavioral trends, highlighting differences in sleep across the week.  


```python
import pandas as pd

# Load the daily activity and sleep data.
df1 = pd.read_csv(r"C:\Users\janan\Downloads\dailyActivity_merged.csv")
df18 = pd.read_csv(r"C:\Users\janan\Downloads\sleepDay_merged.csv")

# Convert date columns to a standard datetime format.
df1['ActivityDate'] = pd.to_datetime(df1['ActivityDate'], format='%m/%d/%Y')
df18['SleepDay'] = pd.to_datetime(df18['SleepDay'], format='%m/%d/%Y %I:%M:%S %p')

# Rename the 'SleepDay' column to 'ActivityDate' so they have the same name.
df18 = df18.rename(columns={'SleepDay': 'ActivityDate'})

# Now, merge the two DataFrames using the 'on' argument for both columns.
df_merged_sleep_activity = pd.merge(df1, df18, on=['Id', 'ActivityDate'], how='inner')

# Display the first 5 rows of the new, merged DataFrame.
print(df_merged_sleep_activity.head())
```

```python

# Calculate the correlation between total minutes asleep and total steps.
correlation = df_merged_sleep_activity['TotalMinutesAsleep'].corr(df_merged_sleep_activity['TotalSteps'])
print(f"The correlation between TotalMinutesAsleep and TotalSteps is: {correlation:.2f}")

#scatter plot to visualize the relationship.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalSteps', y='TotalMinutesAsleep', data=df_merged_sleep_activity)
plt.title('Total Steps vs. Total Minutes Asleep')
plt.xlabel('Total Steps')
plt.ylabel('Total Minutes Asleep')
plt.grid(True)
plt.show()

# visualize the average sleep time on different days of the week.

df_merged_sleep_activity['Weekday'] = df_merged_sleep_activity['ActivityDate'].dt.day_name()
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_merged_sleep_activity['Weekday'] = pd.Categorical(df_merged_sleep_activity['Weekday'], categories=weekday_order, ordered=True)
avg_sleep_by_weekday = df_merged_sleep_activity.groupby('Weekday')['TotalMinutesAsleep'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Weekday', y='TotalMinutesAsleep', data=avg_sleep_by_weekday)
plt.title('Average Sleep Duration per Weekday')
plt.xlabel('Weekday')
plt.ylabel('Average Minutes Asleep')
plt.grid(axis='y')
plt.show()
```

# Hourly Activity, Calories, and Intensity Analysis  

In this part of the analysis, I explored how **steps, calories, and intensity** change throughout the day.  


## Data Preparation  
- I combined three datasets: hourly steps, hourly calories, and hourly intensities.  
- The `ActivityHour` column was converted into a proper datetime format so the data could be aligned across all three files.  
- After merging, I extracted just the **hour of the day (0–23)** from `ActivityHour` to study trends by time of day.  


## Aggregation  
- I grouped the merged data by **hour of the day**.  
- For each hour, I calculated the **average number of steps, average calories burned, and average intensity** across all users.  
- This gave a clear picture of typical daily patterns.  

## Visualization  
- I created three line plots:  
  1. **Average Steps per Hour** – showing when people are most active.  
  2. **Average Calories per Hour** – showing when energy expenditure is highest.  
  3. **Average Intensity per Hour** – reflecting the level of physical exertion throughout the day.  
- All three plots share the same x-axis (hour of the day), making it easy to compare patterns.  


## Insights  
- **Steps**: Activity levels are generally low during late night and early morning, then increase during daytime hours, with peaks typically around **morning commutes and evening exercise times**.  
- **Calories**: The calorie burn trend closely follows the step trend, since calories burned are directly linked to physical movement.  
- **Intensity**: Intensity is low during rest hours and higher during active daytime hours, showing clear spikes when people are most engaged in physical activity.  


## Key Takeaways  
- This analysis highlights **daily activity patterns** across users.  
- Most people are least active during the night and peak in activity during **morning and evening hours**.  
- Understanding these patterns is useful for identifying the **best times for workouts**, energy expenditure, or even personal health recommendations.  


```python

# Load the hourly data files
df_hourly_steps = pd.read_csv(r"C:\Users\janan\Downloads\hourlySteps_merged.csv")
df_hourly_calories = pd.read_csv(r"C:\Users\janan\Downloads\hourlyCalories_merged.csv")
df_hourly_intensities = pd.read_csv(r"C:\Users\janan\Downloads\hourlyIntensities_merged.csv")

# Convert 'ActivityHour' column to a proper datetime format for merging
df_hourly_steps['ActivityHour'] = pd.to_datetime(df_hourly_steps['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
df_hourly_calories['ActivityHour'] = pd.to_datetime(df_hourly_calories['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
df_hourly_intensities['ActivityHour'] = pd.to_datetime(df_hourly_intensities['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')

# Merge the datasets on 'Id' and 'ActivityHour'
df_hourly_merged = pd.merge(df_hourly_steps, df_hourly_calories, on=['Id', 'ActivityHour'], how='inner')
df_hourly_merged = pd.merge(df_hourly_merged, df_hourly_intensities, on=['Id', 'ActivityHour'], how='inner')

# Extract just the hour of the day for grouping
df_hourly_merged['Hour'] = df_hourly_merged['ActivityHour'].dt.hour

print("Hourly merged dataset created. Displaying the first 5 rows:")
print(df_hourly_merged.head())
```

```python
# Group data by hour to find the average steps, calories, and intensity
hourly_trends = df_hourly_merged.groupby('Hour')[['StepTotal', 'Calories', 'TotalIntensity']].mean().reset_index()

# Create a figure with three subplots to show each trend
fig, axes = plt.subplots(3, 1, figsize=(12, 18), sharex=True)
fig.suptitle('Hourly Activity and Calorie Trends', fontsize=20, y=1.02)

# Plot Average Steps per Hour
sns.lineplot(ax=axes[0], x='Hour', y='StepTotal', data=hourly_trends, marker='o', color='b')
axes[0].set_title('Average Steps per Hour')
axes[0].set_ylabel('Average Steps')
axes[0].grid(True)

# Plot Average Calories Burned per Hour
sns.lineplot(ax=axes[1], x='Hour', y='Calories', data=hourly_trends, marker='o', color='r')
axes[1].set_title('Average Calories Burned per Hour')
axes[1].set_ylabel('Average Calories')
axes[1].grid(True)

# Plot Average Intensity per Hour
sns.lineplot(ax=axes[2], x='Hour', y='TotalIntensity', data=hourly_trends, marker='o', color='g')
axes[2].set_title('Average Intensity per Hour')
axes[2].set_xlabel('Hour of the Day (0-23)')
axes[2].set_ylabel('Average Intensity')
axes[2].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

# Daily Activity Analysis – Steps vs. Calories  

In this analysis, I explored the relationship between **daily steps** and **calories burned** using the daily activity dataset.  

## Objective  
The goal was to see whether there is a clear relationship between the **number of steps taken in a day** and the **calories burned**. Since calories burned are often influenced by physical activity, we expect to see a positive relationship.  


## Visualization  
A scatter plot was created with:  
- **X-axis:** Total Steps taken per day  
- **Y-axis:** Calories burned per day  

Each point in the scatter plot represents one day of data for a user.  


## Insights  
- If the points form an upward trend, this indicates a **positive relationship**: as steps increase, calories burned also increase.  
- If the points are widely scattered, it suggests that while steps contribute to calorie burn, other factors (e.g., workout intensity, heart rate, or non-step activities) may also play a role.  
- Outliers may represent days with **very high calories burned** despite relatively fewer steps, possibly due to activities like cycling, swimming, or high-intensity workouts that don’t involve step counts.  

## Key Takeaway  
This analysis confirms that **steps are a strong indicator of calories burned**, but they are not the only factor. Physical intensity and other forms of activity also significantly contribute to daily calorie expenditure.  


```python
df1 = pd.read_csv(r"C:\Users\janan\Downloads\dailyActivity_merged.csv")

# Create a scatter plot to visualize the relationship between steps and calories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalSteps', y='Calories', data=df1)
plt.title('Total Steps vs. Calories Burned')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

# Daily Sedentary Minutes vs. Calories Burned  

In this analysis, I examined the relationship between **time spent sedentary** and **calories burned** using the daily activity dataset.  


## Objective  
The goal was to understand whether spending more time in sedentary activities (e.g., sitting, resting, or being inactive) has any clear relationship with the number of calories burned in a day.  

## Visualization  
A scatter plot was created with:  
- **X-axis:** Sedentary Minutes per day  
- **Y-axis:** Calories burned per day  

Each point on the plot represents a single day of data for a user.  

## Insights  
- If a clear downward trend exists, it suggests that **more sedentary time is associated with fewer calories burned**.

This analysis shows that while being sedentary generally means lower calorie burn, **calories burned depend on total daily activity**. Sedentary minutes alone are not enough to fully predict calorie expenditure, as even short periods of physical activity can significantly impact calorie burn.

```python
# Load the sleep data into a DataFrame named 'df18'
df18 = pd.read_csv(r"C:\Users\janan\Downloads\sleepDay_merged.csv")

# Convert the 'TotalMinutesAsleep' column to 'TotalHoursAsleep'
df18['TotalHoursAsleep'] = df18['TotalMinutesAsleep'] / 60

# Calculate the overall average sleep duration in hours
average_sleep_hours = df18['TotalHoursAsleep'].mean()

# Define the recommended minimum sleep duration
recommended_sleep = 7

# Create a bar plot to compare the average sleep to the recommendation
plt.figure(figsize=(8, 6))
sns.barplot(x=['Average Sleep', 'Recommended Sleep'], y=[average_sleep_hours, recommended_sleep])
plt.title('Average Sleep Duration vs. Recommended Guidelines')
plt.ylabel('Hours Asleep')

# Add the actual values on top of the bars for clarity
plt.text(x=0, y=average_sleep_hours + 0.1, s=f'{average_sleep_hours:.2f} hrs', ha='center')
plt.text(x=1, y=recommended_sleep + 0.1, s=f'{recommended_sleep:.2f} hrs', ha='center')

plt.ylim(0, 10) # Set a consistent y-axis limit
plt.grid(axis='y')
plt.show()
```
